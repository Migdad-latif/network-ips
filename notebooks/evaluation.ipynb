```python
import os, sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve, average_precision_score
sys.path.insert(0, os.path.join('..', 'src'))
from evaluator import Config, DatasetLoader, Evaluator, MetricsEngine
DATASET_PATH = os.path.join('..','data','cicids2017','Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
print(f'File exists: {os.path.exists(DATASET_PATH)}')
print('Setup complete')
```

## 2. Load Dataset

```python
df = DatasetLoader.load(DATASET_PATH)
print(df['label'].str.strip().value_counts())
```

## 3. Run Evaluation

```python
evaluator = Evaluator()
y_true, y_pred, y_score = evaluator.run(df)
metrics = MetricsEngine.compute(y_true, y_pred, y_score)
MetricsEngine.print_report(metrics)
```

## 4. Confusion Matrix

```python
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(6,5))
fig.patch.set_facecolor('#0d1117')
ax.set_facecolor('#0d1117')
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', ax=ax, annot_kws={'size':14,'color':'white'})
ax.set_title('Confusion Matrix', color='#00ff41')
ax.set_xticklabels(['Benign','Attack'])
ax.set_yticklabels(['Benign','Attack'], rotation=0)
plt.tight_layout()
plt.show()
```

## 5. ROC Curve

```python
fpr_v,tpr_v,_=roc_curve(y_true,y_score)
auc=roc_auc_score(y_true,y_score)
fig,ax=plt.subplots(figsize=(7,5))
ax.plot(fpr_v,tpr_v,color='#00ff41',linewidth=2,label=f'IPS (AUC={auc:.4f})')
ax.plot([0,1],[0,1],color='#447744',linestyle='--',label='Random')
ax.set_title('ROC Curve',color='#00ff41')
ax.set_xlabel('FPR')
ax.set_ylabel('TPR')
ax.legend()
ax.grid(True,alpha=0.3)
plt.tight_layout()
plt.show()
```
